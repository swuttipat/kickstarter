### 2.1  What's the best length of time to run a campaign?

Let roughly look into data 'duration' columns

```python
df.duration.describe()
```

Plot histogram of duraion to see the distribution by separate success and failed projects by color

```python
plt.hist(df[df.status_str == 'success'].duration,bins=20,alpha=0.7,label='success',color='tab:blue')
plt.hist(df[df.status_str == 'failed'].duration,bins=20,alpha=0.5,label='Failed',color='tab:orange')
plt.legend()
plt.xlabel('Duration (Days)')
plt.ylabel('Frequency')
plt.title('Histogram of Project Duration');
```

There seem the blue bar is higher than orange plot in the early range of duration. This can imply that lower duration is highy chance to success of funding. <br>
To be more specific, which periods in days are higest success rate, so we look deeper in data range between 1-55 days

create new columns return the values in range of times

```python
def duration_str(x):
    if 0<x<6:
        return '0-5'
    elif 6<=x<11:
        return '06-10'
    elif 11<=x<16:
        return '11-15'
    elif 16<=x<21 :
        return '16-20'
    elif 21<=x<26:
        return '21-25'
    elif 26<=x<31:
        return '26-30'
    elif 31<=x<36:
        return '31-35'
    elif 36<=x<41:
        return '36-40'
    elif 41<=x<46:
        return '41-45'
    elif 46<=x<51:
        return '46-50'
    elif 51<=x<56:
        return '51-55'
    else:
        return '56+'
    
df['duration2'] = df['duration'].apply(lambda x : duration_str(x))

dur_df = df.groupby(['duration2','status_str']).count().name.reset_index()
dur_df = dur_df.pivot(index='duration2',columns='status_str')
dur_df.columns = ['failed','success']
dur_df['success_rate'] = dur_df.success/(dur_df.success+dur_df.failed)*100
dur_df
```

Find which one is the best for run campaign

```python
dur_df.success_rate.nlargest(10)
```

Plot the data in bar plot

```python
color=['tab:blue']*3+['tab:red']+['tab:blue']*8
dur_df.success_rate.plot(kind='bar',color=color)
plt.title('Success Rate by Duration Period')
plt.xticks(rotation=45)
plt.ylabel('Success Rate (%)')
plt.xlabel('Duration Period (Days)');
```

### 2.3: What type of projects would be most successful at getting funded?
Target to find highest success rate from:
* Category

Create new columns name 'success_rate' to show each type success rate to run the campaign

Plot a bar chart from previose table use column 'succuss_rate'

### 4.  Is there an ideal month/day/time to launch a campaign?

As looking in website https://www.kickstarter.com/ column name 'fund_date' did not refer to date that start the campaign. Therefore we have to create new columns 'start_funded_date' that exactly same as genuine lunch date.

```python
def day_to_sec(day):
    return day*24*60*60
```

```python
df['funded_date_unix'] = df['funded_date'].apply(lambda x: dt.datetime.strptime(x, '%a, %d %b %Y %X -%f').timestamp())
df['duration_unix'] = df['duration'].apply(lambda x: day_to_sec(x))
df['start_funded_date_unix'] = df['funded_date_unix']-df['duration_unix']
df['start_funded_date'] = df['start_funded_date_unix'].apply(lambda x : dt.datetime.fromtimestamp(x).strftime('%a, %d %b %Y %X -%f'))
```

```python
df['month'] = df['start_funded_date'].apply(lambda x: dt.datetime.strptime(x, '%a, %d %b %Y %X -%f').strftime('%B'))
df['day_of_week'] = df['start_funded_date'].apply(lambda x : dt.datetime.strptime(x,'%a, %d %b %Y %X -%f').strftime('%A'))
df['hour'] = df['start_funded_date'].apply(lambda x : dt.datetime.strptime(x,'%a, %d %b %Y %X -%f').strftime('%H'))
```

```python
df[['funded_date','start_funded_date','duration']]
```

Then make bar plot show a success rate by the **Month/ Day/ time** as we need.

```python
color=['tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:red']

mdt_df = df.groupby(['month','status_str']).count().name.reset_index()
mdt_df = mdt_df.pivot(index='month',columns='status_str')
mdt_df.columns = ['failed','success']
mdt_df['success_rate'] = mdt_df.success/(mdt_df.success+mdt_df.failed)*100
mdt_df.sort_values('success_rate',ascending=False,inplace=True)

#print(mdt_df)
mdt_df.sort_values('success_rate')['success_rate'].plot(kind='barh',color=color)

plt.title('Success Rate by Month')
plt.ylabel('')
plt.xlabel('Success Rate (%)')
```

```python
color=['tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:red']

day_of_week_df = df.groupby(['day_of_week','status_str']).count().name.reset_index()
day_of_week_df = day_of_week_df.pivot(index='day_of_week',columns='status_str')
day_of_week_df.columns = ['failed','success']
day_of_week_df['success_rate'] = day_of_week_df.success/(day_of_week_df.success+day_of_week_df.failed)*100
day_of_week_df.sort_values('success_rate',ascending=False,inplace=True)

#print(day_of_week_df)
day_of_week_df.sort_values('success_rate')['success_rate'].plot(kind='barh',color=color)

plt.title('Success Rate by Day of Week')
plt.xlabel('Success Rate (%)')
plt.ylabel('')
```

```python
color= ['tab:blue']*9+['tab:red']+['tab:blue']*14
hour_df = df.groupby(['hour','status_str']).count().name.reset_index()
hour_df = hour_df.pivot(index='hour',columns='status_str')
hour_df.columns = ['failed','success']
hour_df['success_rate'] = hour_df.success/(hour_df.success+hour_df.failed)*100
hour_df.sort_values('success_rate',ascending=False,inplace=True)

#print(hour_df)
hour_df.sort_index(ascending=False)['success_rate'].plot(kind='barh',color=color,figsize=(6,6))

plt.title('Success Rate by Hour')
plt.ylabel('')
plt.xlabel('Success Rate (%)')
```

### Extra

#### Highlight max bar in plot
ref: [Pandas + Matplotlib, Make one color in barplot stand out](https://stackoverflow.com/questions/20394091/pandas-matplotlib-make-one-color-in-barplot-stand-out)

```python
ax = hour_df.sort_index(ascending=False)['success_rate'].plot(kind='barh',figsize=(6,6))
for bar in ax.patches:
    bar.set_facecolor('tab:blue')
highlight = hour_df[hour_df.success_rate == hour_df['success_rate'].max()].index[0]
pos = hour_df.sort_index(ascending=False)['success_rate'].index.get_loc(highlight)
ax.patches[pos].set_facecolor('tab:red')
```
